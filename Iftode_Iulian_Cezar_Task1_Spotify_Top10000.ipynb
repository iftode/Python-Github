# Analiza setului de date „Spotify Top 10000” în Google Colab

## Introducere

Scopul acestei analize este explorarea setului de date „Spotify Top 10000” pentru a identifica tendințe importante din industria muzicală.

Prin această analiză am urmărit:
- identificarea artiștilor cei mai prezenți în dataset;
- evidențierea melodiilor cu cea mai mare popularitate;
- analiza evoluției melodiilor explicite de-a lungul timpului;
- prezentarea rezultatelor într-un mod clar și ușor de înțeles pentru echipă.

Analiza a fost realizată în Google Colab folosind Python și bibliotecile pandas, matplotlib și seaborn.

```python
import kagglehub

# descarcă datasetul
path = kagglehub.dataset_download("joebeachcapital/top-10000-spotify-songs-1960-now")

print("Dataset salvat la:", path)
```

```python
import pandas as pd
import os

# afișează fișierele din folder
files = os.listdir(path)
print(files)
```

```python
# încarcă datasetul
file_path = os.path.join(path, "top_10000_1950-now.csv")

df = pd.read_csv(file_path)

# afișează primele rânduri
df.head()
```

```python
# informații despre dataset
df.info()
```

```python
# numărul total de melodii
numar_melodii = df.shape[0]

# numărul total de artiști diferiți
numar_artisti = df["Artist Name(s)"].nunique()

print("Număr total de melodii:", numar_melodii)
print("Număr total de artiști diferiți:", numar_artisti)
```

### Observație

Datasetul conține 9995 melodii din perioada 1950-prezent.  
Numărul artiștilor diferiți arată cât de divers este setul de date și ne ajută să înțelegem dacă analiza este concentrată pe câțiva artiști sau acoperă o varietate mare de interpreți.

```python
# top 5 melodii după popularitate
top_populare = df.sort_values(by="Popularity", ascending=False)[
    ["Track Name", "Artist Name(s)", "Popularity"]
].head(5)

top_populare
```

### Observație

Melodiile cu cel mai mare scor de popularitate reprezintă piesele care au avut cel mai mare impact asupra utilizatorilor Spotify.  
Aceste valori reflectă tendințele moderne din industria muzicală și preferințele publicului.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# top 10 artiști cu cele mai multe melodii
top_artisti = df["Artist Name(s)"].value_counts().head(10)

print(top_artisti)
```

```python
# dimensiunea graficului
plt.figure(figsize=(12,6))

# grafic barplot
sns.barplot(
    x=top_artisti.values,
    y=top_artisti.index
)

# titlu și etichete
plt.title("Top 10 artiști cu cele mai multe melodii")
plt.xlabel("Număr melodii")
plt.ylabel("Artist")

plt.show()
```

### Observație

Graficul evidențiază artiștii care apar cel mai frecvent în dataset.  
Prezența repetată a unor artiști indică impactul și consistența lor în industria muzicală de-a lungul timpului.

Vizualizarea ajută echipa să observe rapid care artiști domină clasamentele Spotify.

```python
# convertim data în format datetime
df["Album Release Date"] = pd.to_datetime(
    df["Album Release Date"],
    errors="coerce"
)

# extragem anul
df["Year"] = df["Album Release Date"].dt.year

# verificare
df[["Album Release Date", "Year"]].head()
```

```python
# numărul melodiilor explicite pe an
explicit_songs = (
    df[df["Explicit"] == True]
    .groupby("Year")
    .size()
)

explicit_songs.head()
```

```python
# dimensiunea graficului
plt.figure(figsize=(14,6))

# line plot
plt.plot(
    explicit_songs.index,
    explicit_songs.values
)

# titlu și etichete
plt.title("Evoluția numărului de melodii explicite în timp")
plt.xlabel("An")
plt.ylabel("Număr melodii explicite")

plt.grid(True)

plt.show()
```

### Observație

Analiza evidențiază o creștere semnificativă a numărului de melodii explicite în ultimele decenii.

Acest trend sugerează schimbări în stilul muzical, în libertatea artistică și în preferințele publicului modern.  
Platformele digitale precum Spotify au contribuit la promovarea unui conținut mai divers și mai puțin restricționat comparativ cu perioadele anterioare.

Vizualizarea ajută echipa să observe modul în care industria muzicală s-a schimbat de-a lungul timpului.

```python
# eliminăm rândurile fără an
df_valid_years = df.dropna(subset=["Year"]).copy()

# transformăm anul în număr întreg
df_valid_years["Year"] = df_valid_years["Year"].astype(int)

# calculăm deceniul
df_valid_years["Decade"] = (df_valid_years["Year"] // 10) * 10

# verificare
df_valid_years[["Track Name", "Artist Name(s)", "Year", "Decade", "Explicit"]].head()
```

```python
# numărul melodiilor explicite pe deceniu
explicit_by_decade = (
    df_valid_years[df_valid_years["Explicit"] == True]
    .groupby("Decade")
    .size()
)

explicit_by_decade
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    explicit_by_decade.index,
    explicit_by_decade.values,
    marker="o"
)

plt.title("Evoluția melodiilor explicite pe decenii")
plt.xlabel("Deceniu")
plt.ylabel("Număr melodii explicite")
plt.grid(True)

plt.show()
```

### Observație

Graficul arată cum a evoluat numărul melodiilor explicite de-a lungul deceniilor.  
Se observă că în primele decenii numărul pieselor explicite este foarte redus, iar creșterea devine mai vizibilă în perioada modernă.

Această evoluție poate sugera schimbări în stilul muzical, în libertatea de exprimare artistică și în modul în care publicul consumă muzică pe platforme digitale precum Spotify.
